```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.datasets import load_diabetes

from sklearn.svm import SVR
```

```python
diabete = load_diabetes()
X = diabete.data
Y = diabete.target
```

```python
X.shape, Y.shape
```

```python
x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size = 0.3, random_state = 8)
```

```python
linear = SVR(kernel='linear')
poly = SVR(kernel='poly')
rbf = SVR(kernel='rbf')
```

```python
linear.fit(x_train, y_train)
poly.fit(x_train, y_train)
rbf.fit(x_train, y_train)
```

```python
linaer_y_pred = linear.predict(x_test)
poly_y_pred = poly.predict(x_test)
rbf_y_pred = rbf.predict(x_test)
```

```python
print("SVR linaer kernel error")
linear.score(x_test, y_test)
```

```python
print("SVR poly kernel error")
poly.score(x_test, y_test)
```

```python
print("SVR rbf kernel error")
rbf.score(x_test, y_test)
```
